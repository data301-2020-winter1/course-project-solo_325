```python
#Chain the operations 
```

```python
#Combined flight data
df1 = df.groupby('firstseen').count().reset_index()
df1.rename(columns = {'Unnamed: 0' : 'Count'}, inplace = True)

#7 day average of all flights
df1['7 Day Avg'] = df1['Count'].rolling(7).mean().shift(-3)
df1.head()

df1.head()
```

```python
#Data for passenger flights
df2 = df_passenger.groupby('firstseen').count().reset_index()
df2.rename(columns = {'Unnamed: 0' : 'Count'}, inplace = True)

#7 day average of passenger flights
df2['7 Day Avg'] = df2['Count'].rolling(7).mean().shift(-3)
df2.head()

df2.head()
```

```python
#Data for cargo flights
df3 = df_cargo.groupby('firstseen').count().reset_index()
df3.rename(columns = {'Unnamed: 0' : 'Count'}, inplace = True)

#7 day average of cargo flights
df3['7 Day Avg'] = df3['Count'].rolling(7).mean().shift(-3)
df3.head()

df3.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Combined plot
sns.set(rc={'figure.figsize':(20, 11)})
ax = sns.lineplot(data = df1, x = 'firstseen', y = 'Count')

ax.set(xlabel="Date",
       ylabel="# of Flights",
       title="Y-X")


```
